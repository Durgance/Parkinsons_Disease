# Checking for Outliers in the data

```python
def outlier_function(df, col_name):
    first_quartile = np.percentile(np.array(df[col_name].tolist()), 25)
    third_quartile = np.percentile(np.array(df[col_name].tolist()), 75)
    IQR = third_quartile - first_quartile
    
    upper_limit = third_quartile+(3*IQR)
    lower_limit = first_quartile-(3*IQR)
    outlier_count = 0
    
    for value in df[col_name].tolist():
        if (value < lower_limit) | (value > upper_limit):
            outlier_count += 1
    return lower_limit, upper_limit, outlier_count
```

```python
plt.figure(figsize=(30,20))
try:
    for i,col in enumerate(df.columns,1):
        plt.subplot(4,6,i)
        sns.boxplot(y=feature_col[i],x="Target",data=df)
        plt.tight_layout()
        out=outlier_function(df,col)
        if out[2]>0:
            print(f"There are {out[2]} outliers in {col}")
except:
    pass
```

```python
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_auc_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import pickle
```

# Implementing Over Sampling due to data imbalance
<img src="https://media.giphy.com/media/MEA9CV6F6lCVqo6Rni/giphy.gif" width=70%>

```python
from imblearn.combine import SMOTETomek
from collections import Counter
```

```python
df_up=df.drop("Target",axis=1)
```

```python
df_up_val=df_up.values
```

```python
smk=SMOTETomek(random_state=42)
X_res,y_res=smk.fit_resample(df_up_val,y)
```

```python
X_res.shape,y_res.shape
```

```python
print(f"Orignal Dataset Shape {Counter(y)}")
print(f"Resampled Dataset Shape {Counter(y_res)}")
```

```python
df_new=pd.concat([pd.DataFrame(X_res),pd.DataFrame(y_res)],axis=1)
```

```python
feature_col.append("Target")
df_new.columns=feature_col
```

```python
df_new.head()
```

# Machine Learning Model at work üñ•Ô∏è

## Using Logistic Regression

```python
kf=model_selection.StratifiedKFold(n_splits=5)
log_acc=[]
for fold , (trn_,val_) in enumerate(kf.split(X=df_new,y=y_res)):
    
    X_train=df_new.loc[trn_,feature_col]
    y_train=df_new.loc[trn_,target_col]
    
    X_valid=df_new.loc[val_,feature_col]
    y_valid=df_new.loc[val_,target_col]
    
    log_scaler=RobustScaler()
    X_train=log_scaler.fit_transform(X_train)
    X_valid=log_scaler.transform(X_valid)

    clf=LogisticRegression()
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_valid)
    print(f"The fold is : {fold} : ")
    print(classification_report(y_valid,y_pred))
    acc=roc_auc_score(y_valid,y_pred)
    log_acc.append(acc)
    print(f"The AUC score for Fold {fold+1} : {acc}")
    #  Saving the model as well as the scaler for each fold
    with open(f"./Logistic/fold_{fold+1}.pkl","wb") as out_file:
        pickle.dump(clf,out_file)
    
    with open(f"./Logistic/scaler_{fold+1}.pkl","wb") as scaler_file:
        pickle.dump(log_scaler,scaler_file)

```

## Using Support Vector Machine Classifier with Linear Kernel

```python
kf=model_selection.StratifiedKFold(n_splits=5)
svm_acc=[]
for fold , (trn_,val_) in enumerate(kf.split(X=df_new,y=y_res)):
    
    X_train=df_new.loc[trn_,feature_col]
    y_train=df_new.loc[trn_,target_col]
    
    X_valid=df_new.loc[val_,feature_col]
    y_valid=df_new.loc[val_,target_col]
    
    svm_scaler=RobustScaler()
    X_train=svm_scaler.fit_transform(X_train)
    X_valid=svm_scaler.transform(X_valid)
    
    clf=SVC(kernel="linear")
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_valid)
    print(f"The fold is : {fold} : ")
    print(classification_report(y_valid,y_pred))
    acc=roc_auc_score(y_valid,y_pred)
    svm_acc.append(acc)
    print(f"The AUC Score for {fold+1} : {acc}")
    #  Saving the model as well as the scaler for each fold
    with open(f"./SVM/fold_{fold+1}.pkl","wb") as out_file:
        pickle.dump(clf,out_file)
    
    with open(f"./SVM/scaler_{fold+1}.pkl","wb") as scaler_file:
        pickle.dump(svm_scaler,scaler_file)
```

## Using K-Nearest Neighbors

```python
kf=model_selection.StratifiedKFold(n_splits=5)
knn_acc=[]
for fold , (trn_,val_) in enumerate(kf.split(X=df_new,y=y_res)):
    
    X_train=df_new.loc[trn_,feature_col]
    y_train=df_new.loc[trn_,target_col]
    
    X_valid=df_new.loc[val_,feature_col]
    y_valid=df_new.loc[val_,target_col]
    
    knn_scaler=RobustScaler()
    X_train=knn_scaler.fit_transform(X_train)
    X_valid=knn_scaler.transform(X_valid)
    
    clf=KNeighborsClassifier(n_neighbors=1)
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_valid)
    print(f"The fold is : {fold} : ")
    print(classification_report(y_valid,y_pred))
    acc=roc_auc_score(y_valid,y_pred)
    knn_acc.append(acc)
    print(f"The AUC Score for {fold+1} : {acc}")
    #  Saving the model as well as the scaler for each fold
    with open(f"./KNN/fold_{fold+1}.pkl","wb") as out_file:
        pickle.dump(clf,out_file)
    
    with open(f"./KNN/scaler_{fold+1}.pkl","wb") as scaler_file:
        pickle.dump(knn_scaler,scaler_file)
```

## Using Random Forest Classifier

```python
kf=model_selection.StratifiedKFold(n_splits=5)
rand_acc=[]
for fold , (trn_,val_) in enumerate(kf.split(X=df_new,y=y_res)):
    
    X_train=df_new.loc[trn_,feature_col]
    y_train=df_new.loc[trn_,target_col]
    
    X_valid=df_new.loc[val_,feature_col]
    y_valid=df_new.loc[val_,target_col]
    
    rf_scaler=RobustScaler()
    X_train=rf_scaler.fit_transform(X_train)
    X_valid=rf_scaler.transform(X_valid)
    
    clf=RandomForestClassifier(n_estimators=55,criterion="entropy")
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_valid)
    print(f"The fold is : {fold} : ")
    print(classification_report(y_valid,y_pred))
    acc=roc_auc_score(y_valid,y_pred)
    rand_acc.append(clf)
    print(f"The accuracy for {fold+1} : {acc}")
    #  Saving the model as well as the scaler for each fold
    with open(f"./RandomForest/fold_{fold+1}.pkl","wb") as out_file:
        pickle.dump(clf,out_file)
    
    with open(f"./RandomForest/scaler_{fold+1}.pkl","wb") as scaler_file:
        pickle.dump(rf_scaler,scaler_file)
```

```python
def best_model():
    
    ## Taking the best models
    with open(f"./Logistic/fold_{np.argmax(log_acc)+1}.pkl","rb") as f:
        log_model = pickle.load(f)
    
    with open(f"./SVM/fold_{np.argmax(log_acc)+1}.pkl","rb") as f:
        svm_model = pickle.load(f)
    
    with open(f"./KNN/fold_{np.argmax(log_acc)+1}.pkl","rb") as f:
        knn_model = pickle.load(f)
    
    with open(f"./RandomForest/fold_{np.argmax(log_acc)+1}.pkl","rb") as f:
        rf_model = pickle.load(f)


    ## Getting the scaler for the same Best accuracy Fold
    with open(f"./Logistic/scaler_{np.argmax(log_acc)+1}.pkl","rb") as f:
        log_scaler = pickle.load(f)
    
    with open(f"./SVM/scaler_{np.argmax(log_acc)+1}.pkl","rb") as f:
        svm_scaler = pickle.load(f)
    
    with open(f"./KNN/scaler_{np.argmax(log_acc)+1}.pkl","rb") as f:
        knn_scaler = pickle.load(f)
    
    with open(f"./RandomForest/scaler_{np.argmax(log_acc)+1}.pkl","rb") as f:
        rf_scaler = pickle.load(f)

    ## Now testing on any part of the data by using test_train_split
    X_train,X_valid,y_train,y_valid=model_selection.train_test_split(df_new[feature_col],df_new["Target"],test_size=0.4)
    

    ## Transforming the Data
    X_valid_log=log_scaler.transform(X_valid)
    X_valid_svm=svm_scaler.transform(X_valid)
    X_valid_knn=knn_scaler.transform(X_valid)
    X_valid_rf=rf_scaler.transform(X_valid)

    ## Getting the prediction value for each model
    y_pred_log=log_model.predict(X_valid_log)
    y_pred_svm=svm_model.predict(X_valid_svm)
    y_pred_knn=knn_model.predict(X_valid_knn)
    y_pred_rf=rf_model.predict(X_valid_rf)

    print(f"Accuracy Score for the LogisticRegression() = {accuracy_score(y_pred_log,y_valid)*100}")
    print(f"Accuracy Score for the SVC(kernel='linear') = {accuracy_score(y_pred_svm,y_valid)*100}")
    print(f"Accuracy Score for the KNeighborsClassifier() = {accuracy_score(y_pred_knn,y_valid)*100}")
    print(f"Accuracy Score for the RandomForestClassifier() = {accuracy_score(y_pred_rf,y_valid)*100}")
    
```
